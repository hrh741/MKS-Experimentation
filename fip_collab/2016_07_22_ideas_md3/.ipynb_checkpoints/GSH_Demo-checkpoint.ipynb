#Generalized Spherical Harmonics Demonstration

The following Jupiter notebook demonstrates the advantages of using Generalized Spherical Harmonics (GSH) as a Fourier basis for functions of orientation. In other words, we can represent any (with certain exceptions) function of orientation with a linear combination of GSH functions.

##Introduction

Orientation is a concept which allows us to define the roll, pitch and yaw of an aircraft or the rotational position of a crystal lattice in space. In this demo we will use Bunge Euler angle to represent orientation, given by $\phi1$, $\Phi$ and $\phi2$.

Sometimes we need to represent a function that depends on orientation; for example, the yield strength of an $\alpha$-titanium single crystal. One simple way to represent a function of orientation is to divide orientation space into discrete bins, and find the average value of our function in each bin. This representation uses a Fourier series where the basis function is simply the indicator function, $\chi_n(\theta)$:

$f(\phi1, \Phi, \phi2) = \sum\limits_p\sum\limits_q\sum\limits_r C_{pqr}\chi_p(\phi1)\chi_q(\Phi)\chi_r(\phi2)$

Unfortunately, since Euler space is large this type of representation requires many coefficients and associated basis functions. The solution is to use a more sophisticated Fourier basis, the set of GSH functions given by,

$T_{l}^{mn}(\phi_1,\Phi,\phi_2)=e^{im\phi_1}P_{l}^{mn}(\Phi)e^{im\phi_2}$

where $P_{l}^{mn}(\Phi)$ are the associated Legendre polynomial functions. The GSH functions are orthonormal and periodic in Euler space. Futhermore, the functions can be modified so that they reflect some desired cystal symmetry. Using the symmetrized GSH bases the function of orientation may be represented as follows,

$f(\phi1, \Phi, \phi2) = \sum\limits_l\sum\limits_m\sum\limits_{\nu} C_l^{m\nu}\dot{\ddot{T}}_{l}^{m\nu}(\phi1, \Phi, \phi2)$

This representation requires far fewer coefficients and associated basis functions.

##Case Study

For this demonstration we have developed a test function that depends on the 3 Bunge Euler angles. The test function exhibits hexagonal crystal symmetry, therefore the domain of unique orientations (the fundamental zone) is limited to $0^{\circ}\leq\phi_1<360^{\circ}$, $0^{\circ}\leq\Phi<90^{\circ}$ and $0^{\circ}\leq\phi_2<60^{\circ}$. The demo allows the user to try out the representation of the test function with both the GSH and indicator bases. The user can change parameters which control the number of basis functions used in each representation, and see the effects on the representation quality.
