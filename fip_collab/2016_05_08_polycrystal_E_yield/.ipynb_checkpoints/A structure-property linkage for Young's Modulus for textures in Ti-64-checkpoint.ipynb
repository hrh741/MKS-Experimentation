This example demonstrates the clustering of different textures in the Ti-64 alloy system and presents a structure property linkage for Young's Modulus. Microstructures were synthetically generated using DREAM.3D and properties were extracted from ABAQUS FEM simulations. Simulations were performed for uniaxial stress boundary conditions in the X loading direction.
